## Описание проекта

**Область исследования**:\
Набор исторических данных с 1980 по 2016 годы:
- продажи игр
- оценки пользователей и экспертов
- жанры и платформы

**Задача**:\
Необходимо запланировать рекламную кампанию на 2017 год: выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Структура данных**:
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Пропуски

<a id="section_14"></a>

Найдем количество пропусков в каждом столбце:

```python
print(data.isnull().sum())
```

Tbd означает, что издатель не захотел получать рейтинговую оценку ESRB или игра не планировалась выпускаться для региона NA.\
Найдем количество значений tbd в столбце user_score:

```python
print(data['user_score'].value_counts().get('tbd', 0))
```

Возможная причина пропусков в колонках critic_score, user_score в том, что данные игры с пропусками изначально не были официально зарелижены и размещены на платформах, где происходит оценка и определение рейтинга. \
Данные игры добавлялись на платформы постфактум, по мере расширения каталога, геймпассов и тд.

Т.к. рейтинг ESRB применяется для США и Канады. Поэтому пропуски в rating могут указывать, что игры выпускались на не только для США.

Обрабатываем пропуски:

**Удаление строк:**
- name(2)
- genre(2)
- year_of_release(269)

Причина: немногочисленнности пропусков

**Замена заглушками (-1), (-2) и (-3):**
- critic_score(8578) на (-1)
- остаток user_score(4277) на (-2)
- tbd в user_score (2424) на (-3)

Причина: игры - не такой типовой объект оценки, как, например, квартира/машина и тд. 
Поэтому, кажется, что заполнять любой мерой центральной тенданции - значительно искажать итоговые выводы исследовательского анализа. А удалять пропуски нельзя из-за большого процента данных. Поэтому все пропуски заменяем разными заглушками, чтобы идентифицировать этот тип среди данных.

**Замена альтернативной категорией (other):**
- rating(6766)

Причина: в исследовательском анализе необходимо будет оценить корреляцию рейтинга на продажи. Поэтому имеет смысл выделить отсутствие рейтинга в отдельную категорию, чтобы учесть в итоговой оценке.


Удалим строки:

```python
def remove_missing_rows(data, columns):
    return data.dropna(subset=columns)

data = remove_missing_rows(data, ['name', 'genre', 'year_of_release'])
```

Заполненим пропуски заглушками:

```python
data['critic_score'] = data['critic_score'].fillna(-1)
```

Заменим пробелы и tbd в user_score на заглушки:

```python
data['user_score'] = data['user_score'].fillna(-2)
data['user_score'] = data['user_score'].replace('tbd', -3)
```

Заменим пропуски рейтинга на other:

```python
data['rating'] = data['rating'].fillna('other')
```

```python
print(data.isnull().sum())
```

### Замена типа данных

- year_of_release float64 → int64 (данные должны быть целочисленными)
- user_score object → float64 (данные должны быть чифрой с плавающей точкой)

```python
data['year_of_release'] = data['year_of_release'].astype('int64')
data['user_score'] = data['user_score'].astype('float64')
```

По текущему списку трудно сделать вывод о характерном сроке жизненного цикла платформ.
Для анализа вернемся к полному массиву.

```python
#выбираем уникальные комбинации года выпуска и платформы
unique_data = data[['year_of_release', 'platform']].drop_duplicates()

#собираем эти комбинации в таблицу для подсчета
platform_year_counts = unique_data.groupby('platform').agg({
    'year_of_release':
    'count'
}).reset_index()
platform_year_counts.rename(columns={'year_of_release': 'release_year_count'},
                            inplace=True)
platform_year_counts
```

<a id="section_4"></a>

```python
#посмотрим распределение на графике
plt.figure(figsize=(10, 6))
sns.histplot(platform_year_counts['release_year_count'],
             kde=True,
             color='#789DBC')

#добавим медиану
median_value = platform_year_counts['release_year_count'].median()
plt.axvline(median_value,
            color='red',
            linestyle='--',
            label=f'Медиана: {median_value}')

#настроим заголовка и меток осей
plt.title('Распределение количества лет выпуска по платформам')
plt.xlabel('Количество лет выпуска')
plt.ylabel('Частота')
plt.legend()

plt.show()

#проверим график описательной статистикой
platform_year_counts.describe()
```

<a id="section_7"></a>

```python
#выбираем платформу X360
platform_data = games[games['platform'] == 'X360']

#фильтруем данные для графика влияния отзывов пользователей на продажи (исключаем значения -2 и -3)
user_score_filtered = platform_data[~platform_data['user_score'].isin([-2, -3]
                                                                      )]

#фильтруем данные для графика влияния отзывов критиков на продажи (исключаем значение -1)
critic_score_filtered = platform_data[platform_data['critic_score'] != -1]

#собираем фигуру
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

#строим диаграмму рассеяния для для влияния пользователей на продажи
sns.scatterplot(x='user_score',
                y='total_sales',
                data=user_score_filtered,
                color='#789DBC',
                ax=axes[0])
axes[0].set_title('Влияние отзывов пользователей на продажи (X360)')
axes[0].set_xlabel('Оценка пользователей')
axes[0].set_ylabel('Суммарные продажи, млн')

#строим диаграмму рассеяния для влияния критиков на продажи
sns.scatterplot(x='critic_score',
                y='total_sales',
                data=critic_score_filtered,
                color='#789DBC',
                ax=axes[1])
axes[1].set_title('Влияние отзывов критиков на продажи (X360)')
axes[1].set_xlabel('Оценка критиков')
axes[1].set_ylabel('Суммарные продажи, млн')

plt.tight_layout()
plt.show()

#считаем корреляцию между отзывами и продажами
user_score_corr = user_score_filtered['user_score'].corr(
    user_score_filtered['total_sales'])
critic_score_corr = critic_score_filtered['critic_score'].corr(
    critic_score_filtered['total_sales'])

print(
    f"Корреляция между оценками пользователей и продажами: {user_score_corr:.2f}"
)
print(
    f"Корреляция между оценками критиков и продажами: {critic_score_corr:.2f}")
```

### Влияние на продажи оценок пользователей и критиков для всех платформ

<a id="section_8"></a>

Дополнительно проверим для PS4, XOne, 3DS, насколько верна независимость продаж от отзывов:

```python
#список платформ для анализа
platforms = ['PS4', 'XOne', '3DS']


#сделаем функцию для анализа платформы
def analyze_platform(platform):
    #выбираем данные для платформы
    platform_data = games[games['platform'] == platform]

    #фильтруем данные для графика влияния отзывов пользователей на продажи (исключаем значения -2 и -3)
    user_score_filtered = platform_data[~platform_data['user_score'].
                                        isin([-2, -3])]

    #фильтруем данные для графика влияния отзывов критиков на продажи (исключаем значение -1)
    critic_score_filtered = platform_data[platform_data['critic_score'] != -1]

    #собираем фигуру
    fig, axes = plt.subplots(1, 2, figsize=(14, 7))

    #строим диаграмму рассеяния для влияния пользователей на продажи
    sns.scatterplot(x='user_score',
                    y='total_sales',
                    data=user_score_filtered,
                    color='#789DBC',
                    ax=axes[0])
    axes[0].set_title(f'Влияние отзывов пользователей на продажи ({platform})')
    axes[0].set_xlabel('Оценка пользователей')
    axes[0].set_ylabel('Суммарные продажи, млн')

    #строим диаграмму рассеяния для влияния критиков на продажи
    sns.scatterplot(x='critic_score',
                    y='total_sales',
                    data=critic_score_filtered,
                    color='#789DBC',
                    ax=axes[1])
    axes[1].set_title(f'Влияние отзывов критиков на продажи ({platform})')
    axes[1].set_xlabel('Оценка критиков')
    axes[1].set_ylabel('Суммарные продажи, млн')

    plt.tight_layout()
    plt.show()

    #считаем корреляцию между отзывами и продажами
    user_score_corr = user_score_filtered['user_score'].corr(
        user_score_filtered['total_sales'])
    critic_score_corr = critic_score_filtered['critic_score'].corr(
        critic_score_filtered['total_sales'])

    print(
        f"Корреляция между оценками пользователей и продажами ({platform}): {user_score_corr:.2f}"
    )
    print(
        f"Корреляция между оценками критиков и продажами ({platform}): {critic_score_corr:.2f}"
    )


#анализируем каждую платформу из списка
for platform in platforms:
    analyze_platform(platform)
```

Исходя из полученных результатов:
- однозначно можно сказать, что оценки пользователей не влияют на продажи
- оценки критиков влияние оказывают, но среднего уровня, т.е. однозначной прямой зависимости нет, скорее стоит учитывать как дополнительный фактор.

### Топ-5 платформ в США, Европе и Японии

```python
#определяем топ-5 платформ по продажам
def top_5_platforms_by_region(games, region):
    return games.groupby('platform')[region].sum().nlargest(5).reset_index()


#функция для замены платформ, не входящих в топ-5, на 'Other'
def replace_non_top_platforms(row, top_5_platforms):
    if row['platform'] in top_5_platforms:
        return row['platform']
    else:
        return 'Other'


#объединяем платформы, не вошедшие в топ-5, в категорию "Other"
def add_other_category_platform(games, region_sales):
    top_5_platforms = top_5_platforms_by_region(
        games, region_sales)['platform'].tolist()
    games['platform'] = games.apply(replace_non_top_platforms,
                                    axis=1,
                                    top_5_platforms=top_5_platforms)
    return games.groupby('platform')[region_sales].sum().reset_index()


#строим круговую диаграмму для региона
def plot_pie_chart_platform(ax, data, region_sales, region_name):
    data = add_other_category_platform(data, region_sales)
    ax.pie(data[region_sales],
           labels=data['platform'],
           autopct='%1.1f%%',
           startangle=140,
           colors=[
               '#789DBC', '#ABBA7C', '#FFDDAE', '#C08497', '#8E7DBE',
               '#F4A259', '#4D8B8B', '#A9D1D7', '#F1C6C1', '#E6B8A2', '#C6D9B2'
           ])
    ax.set_title(f'Распределение платформ в {region_name}')


#собираем фигуру для круговых диаграмм
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

#строим круговые диаграммы
plot_pie_chart_platform(axes[0], games.copy(), 'na_sales', 'Северной Америке')
plot_pie_chart_platform(axes[1], games.copy(), 'eu_sales', 'Европе')
plot_pie_chart_platform(axes[2], games.copy(), 'jp_sales', 'Японии')

plt.tight_layout()
plt.show()
```

## Проверка гипотез

### Гипотеза 1
Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

<a id="section_11"></a>

Гипотезы формулируются по следующему принципу:
есть утверждение, которое мы проверяем (рейтинги платформ XOne и PC одинаковые). H0 строится на отрицании этого тезиса (рейтинги платформ XOne и PC НЕ одинаковые = разные).

H0: Средние пользовательские рейтинги платформ XOne и PC разные\
H1: Средние пользовательские рейтинги платформ XOne и PC одинаковые

В данном случае анализируются две независимые выборки. Будем использовать двухвыборочный t-тест для независимых выборок.\
Т.к. у нас не медицинское высокоточное исследование, можем позволить себе взять для alpha значение в 5%.

```python
#вытаскиваем данные для платформ XOne и PC
x_one_scores = games[(games['platform'] == 'XOne')
                     & (~games['user_score'].isin([-2, -3]))]['user_score']
pc_scores = games[(games['platform'] == 'PC')
                  & (~games['user_score'].isin([-2, -3]))]['user_score']

#стат значимость
alpha = 0.05

#проводим тест и выводи мрезультат
results = st.ttest_ind(
    x_one_scores, pc_scores, equal_var=False
)  # equal_var=False используется, если дисперсии могут быть разными
print('p-значение:', results.pvalue)

#проверяем гипотезу
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')

#кросс-чекнем наше решение по гипотезе
print("Средний пользовательский рейтинг для XOne:", x_one_scores.mean())
print("Средний пользовательский рейтинг для PC:", pc_scores.mean())
```

Итог: средние пользовательские рейтинги платформ XOne и PC разные:
- XOne: 6.521428571428572
- PC: 6.2696774193548395

### Гипотеза 2
Средние пользовательские рейтинги жанров Action и Sports разные.

<a id="section_12"></a>

Утверждение, которое мы проверяем (рейтинги жанров Action и Sports разные). H0 строится на отрицании этого тезиса (рейтинги жанров Action и Sports НЕ разные = одинаковые).

H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые\
H1: Средние пользовательские рейтинги жанров Action и Sports разные

Снова анализируются две независимые выборки с помощью двухвыборочного t-теста для независимых выборок.\
Так же за alpha берем значение в 5%, т.к. исследование не относится к категории высокоточных.

```python
#вытаскиваем данные для жанров Action и Sports
action_scores = games[(games['genre'] == 'Action')
                      & (~games['user_score'].isin([-2, -3]))]['user_score']
sports_scores = games[(games['genre'] == 'Sports')
                      & (~games['user_score'].isin([-2, -3]))]['user_score']

#стат значимость
alpha = 0.05

#проводим тест и выводи мрезультат
results = st.ttest_ind(
    action_scores, sports_scores, equal_var=False
)  # equal_var=False используется, если дисперсии могут быть разными
print('p-значение:', results.pvalue)

#проверяем гипотезу
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')

#кросс-чекнем наше решение по гипотезе
print("Средний пользовательский рейтинг для жанра Action:",
      action_scores.mean())
print("Средний пользовательский рейтинг для жанра Sports:",
      sports_scores.mean())
```

Итог: средние пользовательские рейтинги жанров Action и Sports разные:
- Action: 6.837532133676092
- Sports: 5.238124999999999

### Вывод

Проверка гипотез показала:
- средние пользовательские рейтинги платформ XOne и PC разные
- средние пользовательские рейтинги жанров Action и Sports разные
