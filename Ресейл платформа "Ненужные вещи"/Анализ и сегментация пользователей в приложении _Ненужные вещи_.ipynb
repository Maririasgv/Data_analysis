### Когортный анализ (retention rate)

Для анализа нам потребуется определить старт когорт - дату первого посещения приложения для каждого пользователя:

```python
data_first_visit = data.groupby('user_id').agg({'event_time':'min'})\
                         .reset_index().rename(columns={'event_time':'first_visit'})
data_first_visit['first_visit'] = data_first_visit['first_visit'].dt.floor('D')
data_first_visit.head()
```

Присоединим получившийся столбец к основному датасету:

```python
data = data.merge(data_first_visit, on = 'user_id', how = 'left')
data.sample(5)
```

Считаем лайфтайм для каждого события с округлением до дня:

```python
data['lifetime'] = (data['event_time'] - data['first_visit'])
data['lifetime'] = data['lifetime'] / np.timedelta64(1, 'D')
data['lifetime'] = data['lifetime'].astype(int)

data.sample(5)
```

Оценим на графике:

```python
data['lifetime'].hist(figsize=(10, 5), bins=30, color='#789DBC')
plt.xlabel('Время жизни пользователя (дни)')
plt.ylabel('Количество пользователей')
plt.title('Распределение времени жизни пользователей')
plt.show()

data['lifetime'].describe()
```

Видим, что есть большой хвост с с пользователями, у которых лайфтайм больше 10 дней.

Собираем таблицу с удержанием пользователей:

```python
retention = data.pivot_table(index='first_visit',
                                   columns='lifetime',
                                   values='user_id',
                                   aggfunc='nunique')
retention.head()
```

Считаем размеры когорт и соединяем с таблицей удержания пользователей:

```python
#считаем размеры
cohort_sizes = (
    data.groupby('first_visit')
    .agg({'user_id': 'nunique'})
    .rename(columns={'user_id': 'cohort_size'})
)

#объединяем таблицы
retention_merged = cohort_sizes.merge(
    retention, on='first_visit', how='left'
).fillna(0)

retention_merged.head()
```

Собираем финальную таблицу с когортами:

```python
retention_merged = retention_merged.div(retention_merged['cohort_size'],
                                        axis=0).drop(columns=['cohort_size'])

retention_merged = retention_merged.round(3)
retention_merged.head(5)
```

Строим когорты:

```python
#обозначаем фигуру
plt.figure(figsize=(25, 13))  

#строим тепловую карту
ax = sns.heatmap(
    retention_merged.drop(columns=[0]),
    annot=True, 
    fmt=".2%",
    cmap="Blues",
    cbar_kws={'label': 'Коэффициент удержания'}
)

#настраиваем отображение и метки на осях
plt.title("Удержание пользователей", fontsize=16)
plt.xlabel("Время жизни пользователя (дни)", fontsize=12)
plt.ylabel("Дата первого посещения", fontsize=12)
yticks = ax.get_yticks() 
ax.set_yticklabels(
    [
        retention_merged.index[int(tick)].strftime('%Y-%m-%d')
        for tick in yticks
    ],
    rotation=0
)

plt.show()
```

В целом, динамика выглядит типичной для когорт.\
Интересно отметить, что пики удержания наблюдаются у когорт, стартовавших в выходные (например, 2019-10-12 и 2019-10-26).\
Возможно, пользователи более активно вовлекаются в выходные дни, что даёт этим когортам более высокий показатель удержания на начальном этапе;

### Время, проведённое пользователями в приложении

Для приложения по продаже вещей на вторичном рынке сессии, скорее всего, будут короткими — от просмотра объявления до принятия решения о покупке или возврате в список. 

Возможно, если пользователь не взаимодействует с приложением в течение 10-15 минут, сессия считается завершённой. Это объясняется тем, что приложение не требует постоянного внимания, и длительное отсутствие активности может свидетельствовать о начале нового поиска.

Исходя из вышесказанного, возьмем за базовое время сессии пользователя - 15 минут.\
Сначала отсортируем наш датасет по user_id и event_time, чтобы сессии сформировались корректно:

```python
data = data.sort_values(['user_id', 'event_time']).reset_index(drop=True)
```

Разделяем общий таймг каждого пользователя на сессии по 10 минут:

```python
sessions = (data.groupby('user_id')['event_time'].diff()
            > pd.Timedelta('15Min')).cumsum()
data['session_id'] = data.groupby(['user_id', sessions],
                                  sort=False).ngroup() + 1
data.head(30)
```

Да, видим, что теперь сессии "нарезаны" по 15 минут.\
Оценим распределение количества сессий по пользователям:

```python
user_sessions = data.groupby('user_id').agg({'session_id': 'count'})\
                    .sort_values(by='session_id', ascending=False)

user_sessions['session_id'].hist(figsize=(10, 5), bins=30, color='#789DBC')
plt.title('Распределение количества сессий по пользователям')
plt.xlabel('Количество сессий')
plt.ylabel('Количество пользователей')
plt.show()

#статистическое описание данных
user_sessions['session_id'].describe()
```

Мы видим, что есть пользователи, у которых больше 50 сессий и они очень сильно двигают среднее значение и вообще всю статистику. Оценим данные на ящике:

```python
user_sessions = data.groupby('user_id').agg({
    'session_id': 'count'
}).sort_values(by='session_id', ascending=False)

plt.figure(figsize=(10, 6))
user_sessions['session_id'].plot(kind='box', color='#789DBC')
plt.title('Распределение кол-ва сессий по пользователям')
plt.ylabel('Кол-во сессий')
plt.grid(True, color='lightgray', linestyle='-', linewidth=0.5)
plt.show()
```

## Сегментация пользователей

### Определение сегментов

При опреленеии типа сегментации, мы ориентируемся на итоговый результат исследования - рекомендации по **улучшению в интерфейсе**.
Поэтому **сегментировать будем по количесву действий**, а не, например, по времени сессии.

Причины:
- прямая связь с вовлеченностью: количество действий лучше отражает активность пользователей, их вовлеченность в использование приложения и с вероятностью достижения contacts_show;

- фокус на интерфейсе: позволяет выявить узкие места в навигации и сценарии взаимодействия, которые можно улучшить для увеличения конверсии;

- меньше искажений данных: время сессии может включать пассивные периоды (например, пользователь отвлекся), что не отражает реальную активность. К тому же, время сессий в 15 минут мы взяли на свое усмотрение, что уже может влиять на объективность результата;

- гибкость анализа: сегментация по действиям позволяет углубиться в конкретные сценарии и пути пользователей, что важно для улучшения UX.

Для начала, посчитаем кол-во событий на пользователя и оценим распределение данных:

```python
user_actions = data.groupby('user_id').size().reset_index(name='action_count')\
                   .sort_values(by='action_count', ascending=False)
user_actions.describe()
```

Среднее значение активности пользователей составляет около 17 действий, но распределение сильно скошено вправо из-за высокой максимальной активности (478 действий).\
Группировка учитывает медиану (9 действий) и 75-й перцентиль (17 действий), что позволяет лучше сегментировать пользователей на основе их типичного поведения:
- пользователи, совершившие до 9 действий включительно, относятся к нижним 50% распределения;
- пользователи с количеством действий от 10 до 17 находятся в интервале между 50-м и 75-м перцентилями;
- пользователи, совершившие более 17 действий, входят в верхние 25% активности.

Обозначим границы перцентилей для дальнейшей сегментации:

```python
median_actions = np.percentile(user_actions['action_count'], 50)
upper_quartile_actions = np.percentile(user_actions['action_count'], 75)
print(f"Медианное количество действий: {median_actions}")
print(f"75-й перцентиль количества действий: {upper_quartile_actions}")
```

Сегментируем пользователей на группы:

```python
low_activity_users = user_actions[user_actions['action_count'] <=
                                  median_actions]
print(f"Пользователей с низкой активностью: {len(low_activity_users)}")

high_activity_users = user_actions[user_actions['action_count'] >=
                                   upper_quartile_actions]
print(f"Пользователей с высокой активностью: {len(high_activity_users)}")

mid_activity_users = user_actions[
    (user_actions['action_count'] > median_actions)
    & (user_actions['action_count'] < upper_quartile_actions)]
print(f"Пользователей со средней активностью: {len(mid_activity_users)}")
```

Проверяем корректность распределения пользователей по сегментам и неповторяемость пользователей:

```python
#распределение
total_users = len(user_actions)
segmented_users = len(low_activity_users) + len(mid_activity_users) + len(
    high_activity_users)

if total_users == segmented_users:
    print(
        f"Сегментация выполнена корректно: все {segmented_users} пользователя распределены."
    )
else:
    print(
        f"Ошибка в сегментации: распределено {segmented_users}, а всего пользователей {total_users}."
    )

#пользователь попал только в одну группу
intersection_low_mid = pd.merge(low_activity_users,
                                mid_activity_users,
                                on='user_id',
                                how='inner')
intersection_mid_high = pd.merge(mid_activity_users,
                                 high_activity_users,
                                 on='user_id',
                                 how='inner')
intersection_low_high = pd.merge(low_activity_users,
                                 high_activity_users,
                                 on='user_id',
                                 how='inner')

if len(intersection_low_mid) == 0 and len(intersection_mid_high) == 0 and len(
        intersection_low_high) == 0:
    print(
        "Пользователи не пересекаются между группами, сегментация уникальна.")
else:
    print("Обнаружены пересечения между группами пользователей!")
```

### Какой сегмент чаще возвращаться в приложение

Создаем отдельные датафреймы для каждого сегмента:

```python
#сегментируем
low_activity_ids = low_activity_users['user_id'].unique()
mid_activity_ids = mid_activity_users['user_id'].unique()
high_activity_ids = high_activity_users['user_id'].unique()

#создаем датафреймы
df_low_activity = data.query('user_id in @low_activity_ids')
df_mid_activity = data.query('user_id in @mid_activity_ids')
df_high_activity = data.query('user_id in @high_activity_ids')

print("Пример данных для низкой активности:")
display(df_low_activity.sample(5))
```

Создадим сводные таблицы удержания:

```python
pivot_low_activity = df_low_activity.pivot_table(
    index='first_visit', columns='lifetime', values='user_id', aggfunc='nunique'
)

pivot_mid_activity = df_mid_activity.pivot_table(
    index='first_visit', columns='lifetime', values='user_id', aggfunc='nunique'
)

pivot_high_activity = df_high_activity.pivot_table(
    index='first_visit', columns='lifetime', values='user_id', aggfunc='nunique'
)


print("Сводная таблица удержания для низкой активности:")
display(pivot_low_activity.tail())
```

Вычисляем размеры когорт:

```python
cohort_size_low = df_low_activity.groupby('first_visit').agg({
    'user_id':
    'nunique'
}).rename(columns={'user_id': 'cohort_size'})

cohort_size_mid = df_mid_activity.groupby('first_visit').agg({
    'user_id':
    'nunique'
}).rename(columns={'user_id': 'cohort_size'})

cohort_size_high = df_high_activity.groupby('first_visit').agg({
    'user_id':
    'nunique'
}).rename(columns={'user_id': 'cohort_size'})


print("Размеры когорт для низкой активности:")
display(cohort_size_low.head(10))
```

Объединим размеры когорт и сводные таблицы удержания:

```python
merged_retention_low = cohort_size_low.merge(pivot_low_activity,
                                             on='first_visit',
                                             how='left').fillna(0)
merged_retention_mid = cohort_size_mid.merge(pivot_mid_activity,
                                             on='first_visit',
                                             how='left').fillna(0)
merged_retention_high = cohort_size_high.merge(pivot_high_activity,
                                               on='first_visit',
                                               how='left').fillna(0)


print("Объединенная таблица удержания для низкой активности:")
display(merged_retention_low.tail())
```

Рассчитываем коэффециенты удержания:

```python
normalized_retention_low = merged_retention_low.div(
    merged_retention_low['cohort_size'], axis=0).drop(columns=['cohort_size'])
normalized_retention_low = normalized_retention_low.round(3)

normalized_retention_mid = merged_retention_mid.div(
    merged_retention_mid['cohort_size'], axis=0).drop(columns=['cohort_size'])
normalized_retention_mid = normalized_retention_mid.round(3)

normalized_retention_high = merged_retention_high.div(
    merged_retention_high['cohort_size'], axis=0).drop(columns=['cohort_size'])
normalized_retention_high = normalized_retention_high.round(3)

print("Нормализованная таблица удержания для низкой активности:")
display(normalized_retention_low.head())
```

Строим тепловую карту удержания пользователей с низкой активностью:

```python
plt.figure(figsize=(25, 13))

ax = sns.heatmap(
    normalized_retention_low.drop(columns=[0]),
    annot=True,
    fmt=".2f",
    cmap="Blues",
    cbar_kws={'label': 'Коэффециент удержания'}
)

plt.title("Удержание пользователей с низкой активностью", fontsize=16)
plt.xlabel("Время жизни пользователя (дни)", fontsize=12)
plt.ylabel("Дата первого посещения", fontsize=12)
yticks = ax.get_yticks()
ax.set_yticklabels(
    [
        normalized_retention_low.index[int(tick)].strftime('%Y-%m-%d')
        for tick in yticks
    ],
    rotation=0
)

plt.tight_layout()
plt.show()
```

Строим тепловую карту удержания пользователей со средней активностью:

```python
plt.figure(figsize=(25, 13))

ax = sns.heatmap(
    normalized_retention_mid.drop(columns=[0]),
    annot=True,
    fmt=".2f",
    cmap="Blues",
    cbar_kws={'label': 'Коэффециент удержания'}
)

plt.title("Удержание пользователей со средней активностью", fontsize=16)
plt.xlabel("Время жизни пользователя (дни)", fontsize=12)
plt.ylabel("Дата первого посещения", fontsize=12)
yticks = ax.get_yticks()
ax.set_yticklabels(
    [
        normalized_retention_mid.index[int(tick)].strftime('%Y-%m-%d')
        for tick in yticks
    ],
    rotation=0
)

plt.tight_layout()
plt.show()
```

Посчитаем количество пользователей в каждом сегменте, совершивших целевое действие:

```python
#зададим целевое действие переменной
target_action = 'contacts_show'
target_users_low = df_low_activity[df_low_activity['event_name'] ==
                                   target_action]['user_id'].nunique()
target_users_mid = df_mid_activity[df_mid_activity['event_name'] ==
                                   target_action]['user_id'].nunique()
target_users_high = df_high_activity[df_high_activity['event_name'] ==
                                     target_action]['user_id'].nunique()
```

Расчитаем конверсию в целевое действие для каждого сегмента:

```python
conversion_rate_low = round((target_users_low / unique_users_low) * 100, 2)
conversion_rate_mid = round((target_users_mid / unique_users_mid) * 100, 2)
conversion_rate_high = round((target_users_high / unique_users_high) * 100, 2)
```

Соберем результаты в общем датафрейме:

## Проверка статистических гипотез

### Гипотеза 1
Пользователи из разных источников имеют разную конверсию в просмотры контактов.

Сформулируем нулевую и альтернативную гипотезы:

**H0**: пользователи, пришедшие из разных источников, не демонстрируют значимых различий в конверсии в просмотры контактов;\
**H1**: пользователи, пришедшие из разных источников, демонстрируют значимые различия в конверсии в просмотры контактов.

Для проверки данной гипотезы мы будем использовать z-тест:
- работаем с долями, а z-тест — это стандартный метод для проверки гипотез о разнице в долях между независимыми выборками;
- данный тест позволяет нам оценить, насколько вероятно, что различия между пропорциями наблюдаются случайно, если нулевая гипотеза верна.

Подготовим данные: соберем в отдельные датафреймы пользователей по каждому источнику:

```python
yandex_df = data.loc[data['source'] == 'yandex']
google_df = data.loc[data['source'] == 'google']

display(yandex_df.sample(3))
display(google_df.sample(3))
```

Находим количество уникальных пользователей и их конверсию в целевое действие (contacts_show) для источника yandex:

```python
#зададим целевое действие переменной
target_action_source = 'contacts_show'

yandex_total = yandex_df['user_id'].nunique()
yandex_cr = yandex_df.loc[yandex_df['event_name'] ==
                          target_action_source]['user_id'].nunique()

print('Всего в выборке yandex:', yandex_total)
print('Конверсия у пользователей из yandex:',
      round((yandex_cr / yandex_total) * 100, 1), '%')
```

Находим количество уникальных пользователей и их конверсию в целевое действие (contacts_show) для источника yandex:

```python
google_total = google_df['user_id'].nunique()
google_cr = google_df.loc[google_df['event_name'] ==
                          target_action_source]['user_id'].nunique()

print('Всего в выборке google:', google_total)
print('Конверсия у пользователей из google',
      round((google_cr / google_total) * 100, 1), '%')
```

Соберем данные для z-теста:

Разница между источниками минимальна. Нам важно снизить вероятность ошибочного отклонения нулевой гипотезы, поэтому значение уровня значимости (α) возьмем 0.01.

Хорошо бы усилить тест параметром мощности, но для этого нужно понять требуемый объем выборки:

```python
alpha = 0.01  #уровень значимости
power = 0.8   # мощность
p1 = 0.247    # конверсия для Yandex
p2 = 0.244    # конверсия для Google
delta = abs(p1 - p2)  #минимально значимая разница

#z-значения для уровня значимости и мощности
z_alpha = norm.ppf(1 - alpha / 2)  # двусторонний тест
z_beta = norm.ppf(power)

#cредняя конверсия
p_pool = (p1 + p2) / 2

#расчет размера выборки
n = (((z_alpha + z_beta) ** 2 * (p_pool * (1 - p_pool) * 2)) / (delta ** 2)).round()

print('Для проведения усиленного теста требуется', n, 'уникальных пользователей.')
```

Наши выборки гораздо меньше, поэтому будем проводить тест, полагаясь только на уровень стат значимости в 0,01:

```python
counts = np.array([yandex_cr, google_cr])
nobs = np.array([yandex_total,
                 google_total])

z_stat, p_value = proportions_ztest(counts, nobs, alternative='two-sided')
print(f"Z-статистика: {z_stat:.4f}")
print(f"P-значение: {p_value:.4f}")

if p_value < 0.01:
    print("Разница статистически значима.")
else:
    print("Разница не является статистически значимой.")
```

Ожидаемо, стат значимости не обнаружено. \
Поэтому, **мы не можем отвергнуть нулевую гипотезу**.

### Гипотеза 2
Пользователи c сессией от 10 минут демонстрируют более высокую конверсию в просмотры контактов, чем до 10 минут.

Сформулируем нулевую и альтернативную гипотезы:

**H0**: конверсия в просмотры контактов у пользователей с сессией больше 10 минут равна конверсии у пользователей с сессией меньше или равной 10 минутам.

**H1**: конверсия в просмотры контактов у пользователей с сессией больше 10 минут выше, чем у пользователей с сессией  меньше или равной 10 минутам.

Для проведения данного теста, мы так же будем использовать z-тест для пропорций, так как мы сравниваем конверсии (доли).

Подготовим данные, нам необходимо объединить наш основной датафрейм с тем, в котором мы считали время сессий:

```python
data = data.merge(final_sessions, on='session_id', how='left')
data.sample(3)
```

Разделить данные на две группы по длительности сессий и посчитаем уникальных пользователей для каждой:

```python
#разделение
long_sessions = data[data['session_duration'] > 10]
short_sessions = data[data['session_duration'] <= 10]

#уникальные пользователи
unique_users_long = long_sessions['user_id'].nunique()
unique_users_short = short_sessions['user_id'].nunique()

print(
    f'Количество уникальных пользователей с сессией > 10 минут: {unique_users_long}'
)
print(
    f'Количество уникальных пользователей с сессией <= 10 минут: {unique_users_short}'
)
```

Посчитаем пользователей, совершивших целевое действие в каждой группе:

```python
#зададим целевое действие переменной
target_action_time = 'contacts_show'

target_users_long = long_sessions[long_sessions['event_name'] ==
                                  target_action_time]['user_id'].nunique()
target_users_short = short_sessions[short_sessions['event_name'] ==
                                    target_action_time]['user_id'].nunique()

print(
    f'Пользователи, совершившие целевое действие (сессия > 10 минут): {target_users_long}'
)
print(
    f'Пользователеи, совершившие целевое действие (сессия <= 10 минут): {target_users_short}'
)
```

Считаем конверсию:

```python
conversion_rate_long = (target_users_long / unique_users_long) * 100
conversion_rate_short = (target_users_short / unique_users_short) * 100

print(f'Конверсия (сессия > 10 минут): {conversion_rate_long:.2f}%')
print(f'Конверсия (сессия <= 10 минут): {conversion_rate_short:.2f}%')
```

Конверсия ощутимо различается, поэтому будем использовать общепринятый уровень значимости в 0.05:

```python
count = [target_users_long, target_users_short]
nobs = [unique_users_long, unique_users_short]

stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(f'z-статистика: {stat:.2f}')
print(f'p-значение: {p_value:.4f}')

alpha = 0.05
if p_value < alpha:
    print(
        "Отвергаем нулевую гипотезу: конверсия у пользователей с сессией > 10 минут выше."
    )
else:
    print(
        "Не удалось отвергнуть нулевую гипотезу: конверсия у пользователей с сессией > 10 минут не выше."
    )
```

### Вывод

- конверсии между источниками Yandex и Google статистически не различаются. Улучшения на уровне источников не требуются;
- конверсии выше у пользователей с сессиями > 10 минут. Нужно стимулировать более длительные сессии.
